1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
def education_types(education):
    education = ' '.join(education.split()[:2]).lower()
    if 'неоконченное' in education:
        return 'неоконченное высшее'
    elif 'специальное' in education:
        return 'среднее специальное'
    elif 'высшее' in education:
        return 'высшее'
    elif 'среднее' in education:
        return 'среднее'

vacancies_df['Образование'] = vacancies_df['Образование и ВУЗ'].apply(education_types)
vacancies_df.drop('Образование и ВУЗ', axis=1, inplace=True)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def experience_format(raw_experience):
    if raw_experience == 'Не указано':
        return np.nan
    elif type(raw_experience) is str:
        months_counter = 0
        
        # превращаем строку опыта  в список с только нужными для поиска нам элементами
        sliced_experience = raw_experience.split()[2:6]
        
        # добавим флаг, чтобы не искать повторно год, если он был найден ранее
        found_year = False
        
        # в списке находим слово 'год' или 'лет', если нашли, то предыдущий элемент умножаем на 12 и добавляем в общий счетчик
        for word in range(len(sliced_experience)):
            # добавим флаг, чтобы не производил поиск года, если он был найден ранее, и потом искал только месяц
            if not found_year:
                for year in ['лет', 'год']:
                    if year in sliced_experience[word]:
                        months_counter += int(sliced_experience[word - 1]) * 12
                        found_year = True
            # ищем месяц по тому же принципу, что и год
            if 'месяц' in sliced_experience[word]:
                months_counter += int(sliced_experience[word - 1])
        return months_counter
    else:
        return raw_experience

vacancies_df['Опыт работы (месяц)'] = vacancies_df['Опыт работы'].apply(experience_format)
vacancies_df.drop('Опыт работы', axis=1, inplace=True)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# найдем все возможные виды занятости или графика
def feature_split(df_line, feature_types):
    for feature in df_line.split(','):
        if feature.strip() not in feature_types:
            # feature_types намеренное не возвращаем, т.к. это изменяемый тип данных - список
            feature_types.append(feature.strip())

# обойдем колонки 'Занятость' и 'График' и на их основе создадим новые one hot признаки
one_hot_features = ['Занятость', 'График']
for feature_type in one_hot_features:
    feature_types = []
    vacancies_df[feature_type].apply(lambda x: feature_split(x, feature_types))
    
    for feature in feature_types:
        vacancies_df[feature] = vacancies_df[feature_type].apply(lambda x: True if feature in x else False)

vacancies_df.drop(one_hot_features, axis=1, inplace=True)
```

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
exchanges_df = pd.read_csv('ExchangeRates.csv', sep=',')
currencies = {
    'руб.': 'RUB',
    'грн.': 'UAH', 
    'USD': 'USD', 
    'EUR': 'EUR', 
    'бел.руб.': 'BYN', 
    'KGS': 'KGS', 
    'сум': 'UZS', 
    'AZN': 'AZN', 
    'KZT': 'KZT'
}

vacancies_df['Обновление резюме'] = pd.to_datetime(vacancies_df['Обновление резюме']).dt.date
exchanges_df['date'] = pd.to_datetime(exchanges_df['date']).dt.date

vacancies_df['уровень_зарплаты'] = vacancies_df['ЗП'].apply(lambda x: int(x.split()[0]))
vacancies_df['валюта_зарплаты'] = vacancies_df['ЗП'].apply(lambda x: currencies[x.split()[1]])

vacancies_df = vacancies_df.merge(
    exchanges_df, how='left', left_on=['валюта_зарплаты', 'Обновление резюме'], right_on = ['currency', 'date']
)

vacancies_df.fillna({'proportion': 1, 'close': 1}, inplace=True)
vacancies_df['ЗП(руб)'] = vacancies_df['уровень_зарплаты'] * vacancies_df['close'] / vacancies_df['proportion']
vacancies_df.drop(['ЗП', 'уровень_зарплаты', 'валюта_зарплаты'], axis=1, inplace=True)

# Удалим так же столбцы из валютного датафрейма, т.к. они не имеют никакого значения в последующей работе
vacancies_df.drop(['currency', 'per', 'date','time', 'close', 'vol', 'proportion'], axis=1, inplace=True)
```

```python
# Создадим фугкцию для постройки гистограмы и боксплота
def hist_box_create(data, feature, plots_info, bins=0, xlabel_rotate=False):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

    #  Дадим тайтл и название оси Х каждому из графиков
    for it in [0, 1]:
        axes[it].set_title(plots_info['title'])
        axes[it].set_xlabel(plots_info['xlabel'])
    
    # Создадим название оси Y у левого графика, который будет валиден и для правого
    axes[0].set_ylabel(plots_info['ylabel'])
    
    # Добавим аргументы функции в словарь для того, чтобы не создавать условия if/else при создании гистограммы с бинами или без
    args = {'data': data, 'x': feature}   

    # Cоздадим боксплот
    boxplot = sns.boxplot(**args, ax=axes[1])
    
    if bins:
        args['bins'] = bins
    
    # Cоздадим гистограмму
    histplot = sns.histplot(**args, ax=axes[0])
    
    # Повернем значения по оси Х, если они перекрывают друг друга
    if xlabel_rotate:
        histplot.tick_params(rotation=30)
        boxplot.tick_params(rotation=30)

    
hist_box_create(
    data=vacancies_df, 
    feature='Возраст', 
    plots_info={'title': 'Возрастное распределение соискателей', 'xlabel': 'Возраст', 'ylabel': 'Количество Соискателей'}, 
    bins=45
)
```

Есть ли аномалии для признака ЗП? Обратите внимание на гигантские размеры желаемой заработной платы.
*Совет: постройте гистограмму и коробчатую диаграмму рядом.*


```python
# Функция определения аномалий по методу Тьюри. Логарифмирование признака намеренно не включено, т.к. нам надо видеть
# реальные значения на графиках, чтобы определять зависимости

def anomalies_count(data, feature):
    x = data[feature]
    
    # определим первый, третий квартиль, а так же межквартильных размах
    q_25 = np.quantile(x, 0.25)
    q_75 = np.quantile(x, 0.75)
    IQR = q_75 - q_25
    
    # определим нижнюю и верхнюю границу
    lower_bound = q_25 - (IQR * 1.5)
    upper_bound = q_75 + (IQR * 1.5)

    # определим выбросы
    outers = data[(x < lower_bound) | (x > upper_bound)]
    
    # очистим датасет
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    
    return outers, cleaned 

outers, cleaned = anomalies_count(vacancies_df, 'ЗП(руб)')


for dataset in [vacancies_df, cleaned]:
    hist_box_create(
        data=dataset, 
        feature='ЗП(руб)',
        plots_info={
            'title': 'Распределение по запрашиваемой зарплате', 'xlabel': 'ЗП(руб)', 'ylabel': 'Количество Соискателей'
        },
        xlabel_rotate=True
    )
```

Выводы:
Первые два графика показывают нам что максимально запрашиваемая зарплата составляет ~ 25 млн. рублей, что является выбросом
По этому избавимся от аномалий с помощью метода Тьюки, чтобы графики стали более удобочитаемые. Для этого намеренно не будем включать логирование в наш метод, чтобы не искажать реальный размер запрашиваемых зарплат.
Теперь согласно графика мы можем увидеть, что минимально запрашиваемая зарплата  составляет около рубля(что скорее всего 
является) аномалией. В то же время максимальный уровень запрашиваемой зарплаты равняется около ~150 тысяч. При это медианная
зарплата находится на уровне около 50 тысяч, основной диапазон запрашиваемых зарплат находится в пределах 28-78 тысяч. Аномалии 
в признаке присутствуют, по этому я от них и пытался избавиться.

```python
# разобъем искомые профессии на категории, чтобы попытаться определить зависимость запрашиваемой зарплаты от определенной
# сферы деятельности

# Сформируем упорядоченный массив для наших категорий профессий, чтобы при поиске определенных специальностей не происходила 
# накладка. Например менеджер по недвижимости попал в категорию специалистов по недвижимости, а не менеджеров.
from collections import OrderedDict

job_categories = OrderedDict([
    ('Руководители', ('директор', 'начальник', 'руководитель', 'senior', 'head', 'cio')),
    ('IT',(
            'програмист', 'программист', 'разработчик', 'dev', 'web', 'developer', 'software engineer', 'qa', 'тестировщик',
            'data scientist', 'c/c++', 'engineer'
        )
    ),
    ('Недвижимость', ('едвижимост',)),
    ('Продажи', ('sale', 'продаж')),
    ('it менеджмент', ('manager', 'owner')),
    ('сис админы', ('истемный администратор',)),
    ('инженеры', ('инженер',)),
    ('аналитики', ('аналитик',)),
    ('менеджер проектов', ('менеджер проекта', 'менеджер проектов')),
    ('другие менеджеры', ('менеджер',)),
    ('IT спец', ('специалист по it', 'it-специалист', 'ит-специалист', 'cпециалист it', 'cпециалист ит')),
    ('тех. специалисты',  ('ехническ',)),
    ('интернет маркетинг', ('интернет-маркетолог',)),
    ('монтажники', ('монтажник',))
])

# определяем категории специальности согласно того найдется определенная подстрока в искомой должности
def jobs_categories(job):   
    for key, value in job_categories.items():
        for profession in value:
            if profession in job.lower():                
                return key
        
    return 'другие'

# Добавим категрию профессий  в наш датасет
vacancies_df['категории профессий'] = vacancies_df['Ищет работу на должность:'].apply(jobs_categories)

# Посмотрим распределение искомых зарплат согласно категорий профессий
plot_area_prepare(12, 5)
bar_plot = sns.barplot(
    data=vacancies_df,
    x='категории профессий',
    y='ЗП(руб)'
)
bar_plot.set(title='Распределение запрашиваемой зарплаты согласно специальности')
bar_plot.tick_params(rotation=30)

# посмотрим количество соискателей по категориям
print(100 * '_', '\nРаспределение соискателей по категориям:')
display(vacancies_df['категории профессий'].value_counts())

# посмотрим на распределений по специальностям из категории 'другие'
print(100 * '_', '\nРаспределение профессий из категории "другие"')
display(
    vacancies_df[vacancies_df['категории профессий']=='другие']['Ищет работу на должность:'].value_counts()
)
print(100 * '_')

# Удалим созданную колонку, т.к. задача была только построить график  и проанализировать зависимости, а не создавать новые 
# признаки
vacancies_df.drop('категории профессий', axis=1, inplace=True)
```

В последнем  барплоте я попытался разбить профессии кандидатов на категории. Смысл разбиения заключается в поиске определенной
подстроки в искомой профессии, а потом определение кандидата в категорию согласно результатам поиска.

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# Создадим функцию поиска выбросов методом z-отклонений
def z_score_outliers(df, feature, left_limit=3, right_limit=3, log_feature=False):
    
    # Прологарифмируем признак, если установлен соответсвующий флаг
    feature = np.log(df[feature] + 1) if log_feature else df[feature]
    
    # определим мю и сигму
    mean_value = feature.mean()
    sigma = feature.std()
    
    # определим левую и правую границу согласно формуле z-отклонений
    left_bond = mean_value - sigma * left_limit
    right_bond = mean_value + sigma * right_limit
    
    # найдем выбросы
    outliers = df[(feature < left_bond) | (feature > right_bond)]
    
    # очистим датасет от выбросов
    cleaned = df[(feature > left_bond) & (feature < right_bond)]
    
    return outliers, cleaned, left_bond, right_bond, feature

outliers, cleaned, left_bond, right_bond, log_age = z_score_outliers(
    df=vacancies_df, feature='Возраст', right_limit=4, log_feature=True
)

display(outliers)

plot_area_prepare(5,4)
plt.axvline(log_age.mean(), color='k', lw=4)
plt.axvline(left_bond, color='k', lw=4)
plt.axvline(right_bond, color='k', lw=4)
sns.histplot(
    data = log_age,
    bins=65
).set(title='Распределение по возрасту', ylabel='Количество кандидатов')


vacancies_df.drop(outliers.index, inplace=True)


```
